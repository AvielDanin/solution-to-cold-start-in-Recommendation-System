## cosine similarity dataframe preparation function

```python
# define a function to find the top 5 values in a row
def top_5(row):
    return row.nlargest(5).index.tolist()

def prepDatasets_images(test_products,train_products):
    train_df = images_df.loc[train_products]
    train_mat = train_df.to_numpy()
    test_df = images_df.loc[test_products]
    test_mat = test_df.to_numpy()
    # calculating Cosine similarity of all imagaes in test against all images in train
    a = test_mat.dot(train_mat.T)
    b = (norm(test_mat,axis=1,keepdims=True)*norm(train_mat.T,axis=0,keepdims=True))
    cosine_sim = a/b
    # creating a cosine similarity dataframe
    cosine_sim_df = pd.DataFrame(cosine_sim,columns=train_df.index)
    cosine_sim_df['prod_id'] = test_df.index
    cosine_sim_df.set_index('prod_id',inplace=True)

    #adding the maximal similarity column
    cosine_sim_df['max_sim'] = cosine_sim_df.idxmax(axis=1)   #max.index(axis=1)

    # # apply the function to each row to create the new column
    cosine_sim_df['top_5_prod'] = cosine_sim_df[cosine_sim_df.columns[0:-1]].apply(top_5, axis=1)

    return cosine_sim_df

def prepDatasets_desctiption(test_products,train_products):
    train_df = desctiption_emb.loc[train_products]
    train_mat = train_df.to_numpy()
    test_df = desctiption_emb.loc[test_products]
    test_mat = test_df.to_numpy()
    # calculating Cosine similarity of all imagaes in test against all images in train
    a = test_mat.dot(train_mat.T)
    b = (norm(test_mat,axis=1,keepdims=True)*norm(train_mat.T,axis=0,keepdims=True))
    cosine_sim = a/b
    # creating a cosine similarity dataframe
    cosine_sim_df = pd.DataFrame(cosine_sim,columns=train_df.index)
    cosine_sim_df['prod_id'] = test_df.index
    cosine_sim_df.set_index('prod_id',inplace=True)

    #adding the maximal similarity column
    cosine_sim_df['max_sim'] = cosine_sim_df.idxmax(axis=1)   #max.index(axis=1)

    # # apply the function to each row to create the new column
    cosine_sim_df['top_5_prod'] = cosine_sim_df[cosine_sim_df.columns[0:-1]].apply(top_5, axis=1)

    return cosine_sim_df

```

## Evaluation Functions

```python
def IoU(a, b):
    intersection = list(set(a) & set(b)) 
    union = set(a + b)
    if (len(union)>0):
        return (len(intersection)/len(union))
    return 0

IoU([1,2,3,4,8], [1,2,5,6,7,3,4])

def atLeastOneShot(a, b):
    intersection = list(set(a) & set(b))
    return (len(intersection)>0)

atLeastOneShot([1,2,3,4,8], [1,2,5,6,7,3,4])

def numOfSuccesses(a,b):
    intersection = list(set(a) & set(b))
    return (len(intersection))
numOfSuccesses([1,2,3,4,8], [1,2,5,6,7,3,4])
```
